```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y=np.meshgrid(np.arange(-2, 2, 0.04), np.arange(2, -2, -0.04))
```

```python
x
```

```python
y
```

```python
z=x**2+y**2
```

```python
plt.contourf(z)
plt.colorbar()
```

```python
(u,v) = np.gradient(z)
```

```python
plt.contourf(x,y,z)
plt.colorbar()
plt.quiver(x,y,v,u)
```

```python
# x_01=np.arange(-2, 2.4, 0.4)
# y_01=np.arange(2, -2.4, -0.4)
x_01 = np.linspace(-2, 2, 11)
y_01 = np.linspace(-2,2,11)
x_01, y_01=np.meshgrid(x_01,y_01)
z_01 = x_01**2 + y_01**2
u_01,v_01 = np.gradient(z_01)
```

```python
plt.contourf(x_01,y_01,z_01)
plt.colorbar()
plt.quiver(x_01,y_01,-v_01,-u_01)
```

```python
plt.contourf(x_01,y_01,z_01)
plt.colorbar()
plt.quiver(x_01,y_01,-v_01,-u_01,angles='xy',units='xy')
```
